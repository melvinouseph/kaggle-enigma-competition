```
Y = df1['Survival %'].copy()
```

```
X = df1[['People', 'Population', 'Food Availible(Kg)', 'Width(m)', 'Length(m)', 'Months']].copy()
X_test = df2[['People', 'Population', 'Food Availible(Kg)', 'Width(m)', 'Length(m)', 'Months']].copy()
```

```
Y_test = df2['Id'].copy()
```

```
scaler = StandardScaler()
X = scaler.fit_transform(X)
X_test = scaler.transform(X_test)
```

```
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.30)
```

```
model = tf.keras.models.Sequential([
  tf.keras.layers.Input(shape=(6,)),
  #tf.keras.layers.Dropout(0.8),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(1)
])

optimizer = tf.keras.optimizers.SGD(0.01)
model.compile(optimizer=optimizer,
      loss='mse', metrics=['accuracy'])
  #return model
  
r = model.fit(X_train, Y_train, validation_data=(X_val, Y_val), epochs=100)
```

```
print("Train score:", model.evaluate(X_train, Y_train))
```

```
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.legend()

# Plot the accuracy too
plt.plot(r.history['accuracy'], label='acc')
plt.plot(r.history['val_accuracy'], label='val_acc')
plt.legend()
```

```
P = model.predict(X_test)
P = P.flatten()
```

```
P = P.tolist()
id = df2['Id'].to_list()
Y_pred = pd.DataFrame(list(zip(id, P)), 
               columns =['Id', 'Survival %']) 
Y_pred
```
